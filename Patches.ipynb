```python
import cv2
import os

def split_image_into_patches(image_path, patch_size, output_folder):
    # Carregar a imagem
    image = cv2.imread(image_path)
    
    # Obter as dimensões da imagem
    height, width = image.shape[:2]
    
    # Verificar se a pasta de saída existe, se não, criar
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Loop através da imagem e dividir em patches
    for y in range(0, height, patch_size):
        for x in range(0, width, patch_size):
            # Extrair o patch
            patch = image[y:y+patch_size, x:x+patch_size]
            
            # Definir o nome do arquivo
            filename = os.path.join(output_folder, f'patch_{y}_{x}.png')
            
            # Salvar o patch
            cv2.imwrite(filename, patch)

# Exemplo de uso
image_path = "C:/Users/caioe/1111344.jpg"
patch_size = 128
output_folder = 'patches'

split_image_into_patches(image_path, patch_size, output_folder)
```
